```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def func(x):
    return x ** 2

def grad(x):
    return 2 * x
```

```python
D = 5

X = np.linspace(-D, +D, 20)
Y = func(X)

plt.figure(figsize=(16,6))
plt.plot(X, Y, 'r', label='Y(X)')
```

```python
eps = 0.0001

start_point = 5

learning_rate = 0.1

next_point = start_point - learning_rate*start_point
current_point = start_point
x = []
x.append(next_point)

plt.figure(figsize=(16,6))
plt.plot(X,Y, 'r')
while current_point-next_point > eps:
    current_point = next_point
    
    next_point = current_point - learning_rate * grad(current_point)
    x.append(next_point)

X_grad = np.array(x)
plt.plot(X_grad, func(X_grad), '-*g', label='Grad_X')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
